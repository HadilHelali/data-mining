On a testé ces datasets sur plusieurs algorithmes en utilisant __plusieurs critères de comparaison__ :

\begin{align}
  Pourcentage des instances correctement classifiées = \frac{nombre d’instances correctement classifiées}{nombre d’instances totale}
  \end{align}
  <br/>
  \begin{align}
Rappel = \frac{TP}{TP+FN}
\end{align}
  <br/>
\begin{align}
Précision = \frac{TP}{TP+FP}
\end{align}

##### Pour la dataset ‘Iris’ :

* __pourcentage des instances correctement classifiés :__
Le pourcentage des instances correctement classifiés est __maximal__ pour les algorithmes plus proches voisins avec __k=5__ et __10__. Dans ce cas on choisit le __k=5__. <br/>
* __Le rappel :__
Le rappel est égal à __1__ pour presque __tous__ les algorithmes sauf pour l’arbre de décision __(0.98)__. On peut conclure que cette métrique ne peut pas être prise en considération pour le choix du classifieur. <br/>
* __La précision :__
La précision est égale à __1__ pour __tous__ les algorithmes. Cette métrique non plus ne peut pas être utilisée pour choisir un classifieur. <br/>

> On s’appuie donc sur le pourcentage des instances correctement classifiées pour choisir l’algorithme adéquat qui sera dans notre cas l’algorithme du plus proches voisins avec __k=5__.
