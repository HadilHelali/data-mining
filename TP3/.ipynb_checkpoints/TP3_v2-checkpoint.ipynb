```python
# Première Partie ##

from sklearn import datasets

# Charger le jeux de données iris
irisData = datasets.load_iris()

#Afficher les instances
print (irisData.data)

#Afficher les classes
print (irisData.target) 
```

```python
# ----- 2/ a -----

#Afficher les noms des attributs
print(irisData.feature_names)

#Afficher les noms des étiquettes
print(irisData.target_names)

#Afficher le nombre d'instance pour chaque classe
for i in [0,1,2] :
    print(len(irisData.target[irisData.target == i ]))
    
#Afficher le 32ème element
print(irisData.data[31],irisData.target[31])
```

```python
# ----- 2/ b -----
import itertools
import matplotlib
import pylab as pl
def plot_2D(data, target, target_names):
    
    #attribuer une couleur à chaque classe
    colors= itertools.cycle('rgbcmykw') 
    
    #interval [ 0, nombre de classes - 1 ]
    target_ids = range(len(target_names))
    
    #Conteneur des élements de la figure
    pl.figure(facecolor="pink",figsize=(4.8,3.9))
    
    #Dessiner le nuage de points représentant les instances par un couple d'attributs pour chaque classe
    #data[target == i, n] : représente les valeurs de l'énième attribut 
    for i, c, label in zip(target_ids, colors, target_names):
        pl.scatter(data[target == i, 0], data[target == i, 1], c=c, label=label)
        
    #identifier les classes (en haut à droite de la figure)
    pl.legend()
    pl.show()
    
#Appel de la fonction
plot_2D(irisData.data,irisData.target,irisData.target_names)
```

```python
#Même traitement mais avec des couples différents
#Exemple : 2ème attribut en abscisse et le 3ème en ordonnée

def plot_2D_1(data, target, target_names):
    
    colors= itertools.cycle('rgbcmykw') 
    
    target_ids = range(len(target_names))
    
    pl.figure(facecolor="c",figsize=(10,4))
    
    for i, c, label in zip(target_ids, colors, target_names):
        pl.scatter(data[target == 0, 2], data[target == i, 3], c=c, label=label)
        
    pl.legend()
    pl.show()
    
plot_2D_1(irisData.data,irisData.target,irisData.target_names)
```

```python
# Pour dessiner une droite : pl.plot([x1,x2],[y1,y2])
# Il existe une droite qui sépare la classe "setosa" des 2 autres classes dans la question (2/a) 
# L'équation : y = 0.77 * x - 1.03

def plot_2D_seperated(data, target, target_names):
    
    colors= itertools.cycle('rgbcmykw') 
    
    target_ids = range(len(target_names))
    
    pl.figure(facecolor="y",figsize=(4.8,3.9))
    
    #L'équation : y = 0.77 * x - 1.03
    pl.plot([3,7.5],[1.3,4.8],"y")    

    
    for i, c, label in zip(target_ids, colors, target_names):
        pl.scatter(data[target == i, 0], data[target == i, 1], c=c, label=label)
        
    pl.legend()
    pl.show()

plot_2D_seperated(irisData.data,irisData.target,irisData.target_names)
```

```python
## Deuxième Partie ##

# ----- 2/ a -----

# Naive Bayes est implanté en python

from sklearn import naive_bayes

# MultinomialNB : classificateur destiné aux attributs de type discret
# Il en existe plusieurs d'autres, exemple : BernoulliNB qui est destiné aux attributs (booléans / binaires) ...
nb= naive_bayes.MultinomialNB(fit_prior=True) # un algo d'apprentissage

# Entraîner le modèle sur tout le dataset sauf le dernier
nb.fit(irisData.data[:-1], irisData.target[:-1])

#predict classe de 32ème point
p31= nb.predict([irisData.data[31]])
print(p31)

plast= nb.predict([irisData.data[-1]]) 
print(plast)

#prédire la classe de tout le dataset
p= nb.predict(irisData.data[:])
print(p)

print(irisData.target[:])

# Le résultat de prédiction change selon la nature des données et de leurs classes 
# Si on fait varier les instances (les données d'apprentissage) les valeurs des probabilités 
# varient et la classe prédite risque d'être fausse 
```

```python
# ----- 2/ b -----

from sklearn import naive_bayes, datasets

nb = naive_bayes.MultinomialNB(fit_prior=True)

# Charger le jeux de données iris
irisData = datasets.load_iris()

# Entraîner le modèle sur les 99 premières instances 
nb.fit(irisData.data[:99], irisData.target[:99])

# Tester le modèle sur le reste des instances 
P = nb.predict(irisData.data[100:150])
print(P)

# Ce programme entraîne un modèle sur un set composé de 50 instances de la classe 0 et 49 d'autres de la classe 1 
# Il fait le test sur les instances de la classe 2 à qu'elles il donne une prédiction fausse (classe 2 # classe 1)
# parce qu'il ne reconnaît pas la classe 2 puisu'il n'a pas été entraîné sur des instances étiquetées avec la classe 2

#### Solution #### 

    # Il faut toujours faire apparaître toutes les classes possibles dans la phase d'apprentissage 
    # On peut prendre 35 instances de chacune des classes pour l'apprentissage et laisser le reste pour le test
```

```python
## Troixième Partie ##

# ----- 1/ 1 -----

from sklearn import naive_bayes, datasets

nb = naive_bayes.MultinomialNB(fit_prior=True)

# Charger le jeux de données iris
irisData = datasets.load_iris()

# Entrainer le modèle sur les 99 premières instances 
nb.fit(irisData.data[:99], irisData.target[:99])

#########################
#   Première méthode    #
#########################

#nbre d'instances incorrectement classifiés
ea = 0

# Tableau qui contient les prédictions
P = nb.predict(irisData.data[:])

# Tableau qui contient les cibles
Y = irisData.target[:]

for i in range(len(irisData.data)):
    if (P[i] != Y[i]):
        ea = ea+1
        
# Training error (TE)
print(ea/len(irisData.data))

# ----- 1/ 2 -----

#########################
#   Deuxième méthode    #
#########################

# Les valeurs nulles représentent l'égalité entre la prédiction et la classification réelle <=> bonne prédiction
print(P-Y)

# Le nombre des valeurs non nulles
print(len((P-Y)[P-Y != 0]))

# Training error (TE)
print(len((P-Y)[P-Y != 0])/len(irisData.data))


# ----- 1/ 3 -----

#########################
#   Troixième méthode   #
#########################

# Le taux de bonne classification (accuracy)
accuracy = nb.score(irisData.data,Y)

# Training error (TE)
print(1 - accuracy)


# Pour chaque méthode : on a obtenu une erreur égale à 0.3333333 
# L'erreur de prédiction apparaît pour les données de la classe 2 parce que notre modèle ne reconnaît pas la classe
# à cause de l'insuffisance des données d'apprentissage 

```

```python
# ----- 2/ 1 -----

# S1 : Echantillon d'apprentissage , S2 : Echantillon de test 
#|S1| = 2/3 |S| ; |S2| = 1/3 |S|
# [dataS1, targetS1, dataS2, targetS2]

from sklearn import naive_bayes, datasets
import random

# Charger le jeux de données iris
irisData = datasets.load_iris()

def split(data):
    
    initialData = list(data.data)
    initialTarget = list(data.target)
    
    c = list(zip(initialData, initialTarget))
        
    random.shuffle(c)

    initialData, initialTarget = zip(*c)

    return [initialData[:100],initialTarget[:100],initialData[100:],initialTarget[100:]]

split(irisData)
```

```python
def test(S, clf):
    result = split(S)
    clf.fit(result[0],result[1])
    p = clf.predict(result[2])
    accuracy = clf.score(result[2],result[3])
    e = 1 - accuracy
    return e


nb = naive_bayes.MultinomialNB(fit_prior=True)

print(test(irisData,nb))

# L'erreur calculée est toujours inférieur à celle précédemment trouvée 
# parce que cette fois ci on a pris en compte toutes les classes possibles dans la phase d'apprentissage

# Puisque le choix des instances se fait aléatoirement => l'erreur n'est pas toujours la même
    
```

```python
### c.
# t = 10; 50; 100; 200; 500; 1000
def testT(S, clf,it):
    testMoy = 0
    for i in range(0,it):
        testMoy += test(S,clf)
    testMoy /= it
    return testMoy

print(testT(irisData,nb,10))
print("-------------")
print(testT(irisData,nb,50))
print("-------------")
print(testT(irisData,nb,100))
print("-------------")
print(testT(irisData,nb,200))
print("-------------")
print(testT(irisData,nb,500))
print("-------------")
print(testT(irisData,nb,1000))
```

```python
def testT20(S, clf,it):
    testMoy1 = []
    for i in range(0,20):
        testMoy1.append(testT(S,clf,it))
    return testMoy1

print(testT20(irisData,nb,10))
print("-------------")
print(testT20(irisData,nb,50))
print("-------------")
print(testT20(irisData,nb,100))
print("-------------")
print(testT20(irisData,nb,200))
print("-------------")
print(testT20(irisData,nb,500))
print("-------------")
print(testT20(irisData,nb,1000))

# Question : pour chaque t, est-ce que l'erreur moyenne est stable ou non ?
# -> Non , l'erreur moyenne change pour chaque t répété 20 fois
# -> Les tests dépend fortement de la répartition des données qui diffère 
# d'une itération à une autre 
```

```python
from sklearn import naive_bayes, datasets
import random

# Charger le jeux de données iris
irisData = datasets.load_iris()

def splitStabP(data,P):
    initialData = list(data.data)
    initialTarget = list(data.target)

    return [initialData[:int(150*(1-P))],initialTarget[:int(150*(1-P))],initialData[int(150*P):],initialTarget[int(150*P):]]

def testP(S, clf,P):
    result = splitStabP(S,P)
    clf.fit(result[0],result[1])
    p = clf.predict(result[2])
    accuracy = clf.score(result[2],result[3])
    e = 1 - accuracy
    return e

splitStabP(irisData,0.1)
```

```python
### d.

print(testP(irisData,nb,0.1))
sc_Div_Ech = testP(irisData,nb,0.1)
P = float(1/3)
print(testP(irisData,nb,P))
# On remarque que l'erreur estimée dans sa version stable est plus élevé avec 
# un échantillon de test qui ne prend que le 1/3 de l'échantillon initial
# que celui qui ne prend que le 10ème 
```

```python
### e.
from sklearn.model_selection import train_test_split

test_size = [0.33,0.1,0.5]
for i in test_size:
    D_train, D_test, C_train, C_test = train_test_split(irisData.data,irisData.target,test_size=i)
    nb = naive_bayes.MultinomialNB(fit_prior=True)
    nb.fit(D_train, C_train)
    TY = nb.predict(D_test)
    accuracy = nb.score(D_test,C_test)
    e = 1 - accuracy
    print(e)
```

```python
## 2.2 Estimer l'erreur réelle par validation croisée
from sklearn.model_selection import KFold, cross_val_score

nb = naive_bayes.MultinomialNB(fit_prior=True)
k_folds = KFold(n_splits = 10)
scores = cross_val_score(nb, irisData.data, irisData.target, cv = k_folds) 
print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
sc_CV = scores.mean()
```

```python
n_split = [2,3,5,8]
for i in n_split:
    nb = naive_bayes.MultinomialNB(fit_prior=True)
    k_folds = KFold(n_splits = i)
    scores = cross_val_score(nb, irisData.data, irisData.target, cv = k_folds) 
    print("Cross Validation Scores: ", scores)
    print("Average CV Score: ", scores.mean())
    print("--------------------------------------")
```
