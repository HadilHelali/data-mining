````python
import itertools
import matplotlib
import pylab as pl
def plot_2D(data, target, target_names):
    
    #attribuer une couleur à chaque classe
    colors= itertools.cycle('rgbcmykw') 
    
    #interval [ 0, nombre de classes - 1 ]
    target_ids = range(len(target_names))
    
    #Conteneur des élements de la figure
    pl.figure(facecolor="pink",figsize=(4.8,3.9))
    
    #Dessiner le nuage de points représentant les instances par un couple d'attributs pour chaque classe
    #data[target == i, n] : représente les valeurs de l'énième attribut 
    for i, c, label in zip(target_ids, colors, target_names):
        pl.scatter(data[target == i, 0], data[target == i, 1], c=c, label=label)
        
    #identifier les classes (en haut à droite de la figure)
    pl.legend()
    pl.show()
    
#Appel de la fonction
plot_2D(irisData.data,irisData.target,irisData.target_names)
````

````python
#Même traitement mais avec des couples différents
#Exemple : 2ème attribut en abscisse et le 3ème en ordonnée

def plot_2D_1(data, target, target_names):
    
    colors= itertools.cycle('rgbcmykw') 
    
    target_ids = range(len(target_names))
    
    pl.figure(facecolor="c",figsize=(10,4))
    
    for i, c, label in zip(target_ids, colors, target_names):
        pl.scatter(data[target == i, 2], data[target == i, 3], c=c, label=label)
        
    pl.legend()
    pl.show()
    
plot_2D_1(irisData.data,irisData.target,irisData.target_names)
````

## III- Evaluer les performances d'un classiffieur

### 1. Performances sur l'ensemble d'apprentissage

<b> 1. Soit P le tableau qui contient les prédictions, et Y le tableau qui contient les cibles </b><code>(Y = irisData.target)</code><b>. Le code suivant permet de compter les erreurs et ainsi de calculer l'erreur d'apprentissage : </b>

````python
from sklearn import naive_bayes, datasets

nb = naive_bayes.MultinomialNB(fit_prior=True)

# Charger le jeux de données iris
irisData = datasets.load_iris()

# Entrainer le modèle sur les 99 premières instances 
nb.fit(irisData.data[:99], irisData.target[:99])

#########################
#   Première méthode    #
#########################

#nbre d'instances incorrectement classifiés
ea = 0

# Tableau qui contient les prédictions
P = nb.predict(irisData.data[:])

# Tableau qui contient les cibles
Y = irisData.target[:]

for i in range(len(irisData.data)):
    if (P[i] != Y[i]):
        ea = ea+1
        
# Training error (TE)
TE = ea/len(irisData.data)
print("Training Error (TE) : ",ea/len(irisData.data))
````

````python
#########################
#   Deuxième méthode    #
#########################

print("Différence entre la prédiction et la classification réelle : ")
print(P-Y)

# Le nombre des valeurs non nulles
print("Le nombre de valeurs non nulles : ",len((P-Y)[P-Y != 0]))

# Training error (TE)
print("Training Error (TE) : ",len((P-Y)[P-Y != 0])/len(irisData.data))
````

> __remarque :__
> Les valeurs nulles représentent l'égalité entre la prédiction et la classification réelle <=> bonne prédiction


````python
#########################
#   Troixième méthode   #
#########################

# Le taux de bonne classification (accuracy)
accuracy = nb.score(irisData.data,Y)

# Training error (TE)
print("Training Error (TE) : ",1 - accuracy)
````

__Pour chaque méthode : on a obtenu une erreur égale à 0.3333333__ <br/>
L'erreur de prédiction apparaît pour les données de la classe 2 parce que notre modèle ne reconnaît pas la classe à cause de l'insuffisance des données d'apprentissage 

### 2. Performances en généralisation

#### 2.1. Estimer l'erreur réelle par division de l'échantillon d'apprentissage

<b>a. Créez la fonction split(S) pour séparer en deux un échantillon S de couples (X, y) supposés uniformément repartis, selon le principe évoqué précédemment.</b>

 <code>S1</code> : Echantillon d'apprentissage , <code>S2</code> : Echantillon de test <br/><br/>
  \begin{align}
  |S1| = \frac{2}{3} |S| ; |S2| = \frac{1}{3} |S|
  \end{align}<br/>
 <code>[dataS1, targetS1, dataS2, targetS2]</code>

````python
from sklearn import naive_bayes, datasets
import random

# Charger le jeux de données iris
irisData = datasets.load_iris()

def split(data):
    
    initialData = list(data.data)
    initialTarget = list(data.target)
    
    c = list(zip(initialData, initialTarget))
        
    random.shuffle(c)

    initialData, initialTarget = zip(*c)

    return [initialData[:100],initialTarget[:100],initialData[100:],initialTarget[100:]]

split(irisData)
````

<b>b. Créez la fonction </b><code>test(S, clf)</code><b> qui sépare S en deux parties, puis apprend sur la première partie un modèle avec l'estimateur clf, et teste ce modèle sur l'autre partie, en comptant les erreurs faites sur l'échantillon : la proportion d'erreurs est renvoyée par la fonction. Testez cette fonction sur iris avec naive Bayes. L'erreur estimée est-elle plus petite que l'erreur apparente précédemment calculée ? Obtient-on toujours la même estimation pour l'erreur réelle ?</b>

L'erreur calculée est __toujours inférieur__ à celle précédemment trouvée parce que cette fois ci on a pris en compte __toutes les classes possibles__ dans la phase d'apprentissage <br/>
Puisque __le choix des instances__ se fait __aléatoirement__ => l'erreur __n'est pas toujours la même__

````python
def test(S, clf):
    result = split(S)
    clf.fit(result[0],result[1])
    p = clf.predict(result[2])
    accuracy = clf.score(result[2],result[3])
    e = 1 - accuracy
    return e


nb = naive_bayes.MultinomialNB(fit_prior=True)

print("Training Error (TE) : ",test(irisData,nb))

````

<b>c. Réalisez t fois le test de la fonction test sur le même jeu de données et avec le même algorithme d'apprentissage, puis moyennez les erreurs estimées. Quelle erreur moyenne obtenez-vous pour </b><code> t = 10; 50; 100; 200; 500; 1000 </code><b> ? Puis, pour chaque t, répétez 20 fois l'expérience : pour chaque t, est-ce que l'erreur moyenne est stable ou non ? Pouvez-vous interpréter ce résultat ?</b>

````python
def testT(S, clf,it):
    testMoy = 0
    for i in range(0,it):
        testMoy += test(S,clf)
    testMoy /= it
    return testMoy

print("Training Error pour :")
print("t = 10 : ",testT(irisData,nb,10))
print("---------------------------------")
print("t = 50 : ",testT(irisData,nb,50))
print("---------------------------------")
print("t = 100 : ",testT(irisData,nb,100))
print("---------------------------------")
print("t = 200 : ",testT(irisData,nb,200))
print("---------------------------------")
print("t = 500 : ",testT(irisData,nb,500))
print("---------------------------------")
print("t = 1000 : ",testT(irisData,nb,1000))
````

````python
def testT20(S, clf,it):
    testMoy1 = []
    for i in range(0,20):
        testMoy1.append(testT(S,clf,it))
    return testMoy1

print(testT20(irisData,nb,10))
t10 = testT20(irisData,nb,10)
moyt10 = (sum(t10))/len(t10)
print("t = 10 |" ,moyt10 )
print("-------------")
print(testT20(irisData,nb,50))
t50 = testT20(irisData,nb,50)
moyt50 = (sum(t50))/len(t50)
print("t = 50 |" ,moyt50 )
print("-------------")
print(testT20(irisData,nb,100))
t100 = testT20(irisData,nb,100)
moyt100 = (sum(t100))/len(t100)
print("t = 100 |" ,moyt100 )
print("-------------")
print(testT20(irisData,nb,200))
t200 = testT20(irisData,nb,200)
moyt200 = (sum(t200))/len(t200)
print("t = 200 |" ,moyt200 )
print("-------------")
print(testT20(irisData,nb,500))
t500 = testT20(irisData,nb,500)
moyt500 = (sum(t500))/len(t500)
print("t = 500 |" ,moyt500 )
print("-------------")
print(testT20(irisData,nb,1000))
t1000 = testT20(irisData,nb,1000)
moyt1000 = (sum(t1000))/len(t1000)
print("t = 1000 |" ,moyt1000 )
````

> On peut conclure que __les taux d'erreur obtenus__ sont des estimations d'un __même l'estimateur__ (théorique). Plus l'ensemble de données est grand, plus les taux d'erreur __converge vers le taux réel d'erreur__.

<b>d. Est-ce que l'erreur estimée (dans sa version stable), dans le cas d'un échantillon de test qui ne prend que le 10ème de l'échantillon initial, est la même qu'avec la proportion d'1/3 ?</b>

````python
from sklearn import naive_bayes, datasets
import random

# Charger le jeux de données iris
irisData = datasets.load_iris()

def splitStabP(data,P):
    initialData = list(data.data)
    initialTarget = list(data.target)

    return [initialData[:int(150*(1-P))],initialTarget[:int(150*(1-P))],initialData[int(150*P):],initialTarget[int(150*P):]]

def testP(S, clf,P):
    result = splitStabP(S,P)
    clf.fit(result[0],result[1])
    p = clf.predict(result[2])
    accuracy = clf.score(result[2],result[3])
    e = 1 - accuracy
    return e

print("Split du dataset en prenant 10% de l'échantillon initial pour le test : ")
splitStabP(irisData,0.1)
````

````python
print(testP(irisData,nb,0.1))
sc_Div_Ech = testP(irisData,nb,0.1)
P = float(1/3)
print(testP(irisData,nb,P))
````

> On remarque que __l'erreur estimée__ dans sa version stable est __plus élevé__ avec un échantillon de test qui ne prend que le __1/3__ de l'échantillon initial que celui qui ne prend que le __10ème__

<b>f. Testez cette fonction avec plusieurs valeurs pour test size (dont 0.33) pour estimer l'erreur réelle de naïve Bayes sur les données Iris. Quelles erreurs obtenez-vous ?</b>

````python
from sklearn.model_selection import train_test_split

test_size = [0.33,0.1,0.5]
for i in test_size:
    D_train, D_test, C_train, C_test = train_test_split(irisData.data,irisData.target,test_size=i)
    nb = naive_bayes.MultinomialNB(fit_prior=True)
    nb.fit(D_train, C_train)
    TY = nb.predict(D_test)
    accuracy = nb.score(D_test,C_test)
    e = 1 - accuracy
    print("erreur avec test size = ",i," : ",e)
````

#### 2.2 Estimer l'erreur réelle par validation croisée

````python
from sklearn.model_selection import KFold, cross_val_score

nb = naive_bayes.MultinomialNB(fit_prior=True)
k_folds = KFold(n_splits = 10)
scores = cross_val_score(nb, irisData.data, irisData.target, cv = k_folds) 
print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
sc_CV = scores.mean()
````

````python
n_split = [2,3,5,8]
for i in n_split:
    nb = naive_bayes.MultinomialNB(fit_prior=True)
    k_folds = KFold(n_splits = i)
    scores = cross_val_score(nb, irisData.data, irisData.target, cv = k_folds) 
    print("Cross Validation Scores: ", scores)
    print("Average CV Score: ", scores.mean())
    print("--------------------------------------")
````
